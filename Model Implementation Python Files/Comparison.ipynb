```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (18,9)


N = 7
ind = np.arange(7)
width = 0.20

vals1 = [95.542, 97.476, 97.318,98.264,98.107,98.107,98.738]
bar1 = plt.bar(ind, vals1, width, color = 'y')

vals2 = [96.989, 98.039, 97.719,98.511,98.219,98.507,98.507]
bar2 = plt.bar(ind+width, vals2, width, color = 'r')

vals3 = [95.238, 97.402, 97.244,98.365,98.219,98.214,98.809]
bar3 = plt.bar(ind+width*2, vals3, width, color='g')

vals4 = [93.548, 96.774, 96.774,98.219,98.219,97.922,98.516]
bar4 = plt.bar(ind+width*3, vals4, width, color = 'b')



plt.xlabel("Models")
plt.ylabel('Scores')
plt.title("Comparison Of all the 3 Models")
plt.ylim(93,100)

plt.xticks(ind+width*2,['DT', 'RF', 'GB','SVM','KNN','Logistic','Ensemble'])
plt.legend( (bar1, bar2, bar3, bar4), ('Accuracy', 'Precison', 'F1', 'Recall') )
plt.show()
```
